```python
'''
********************************************************************************************************************************
Implementação:

    Foi verificado que os arquivos entregues estão com extensão csv, ao abrir os arquivos foi percebido que o padrão de delimi-
tação para os campos é utilizado o caractere ",". Para melhor visualização da base de dados os arquivos serão convertidos para
arquivos em excel já com os dados estruturados em coluas e com seus valores.

   O arquivo "experiência_trabalho" foi convertido para arquivo de excel, assim já sendo separado automaticamente as colunas e
forma identificada as colunas "company" que possui a empresa em que o colaborador trabalhou, "date" com a informação o tempo de
permanência, "location" com a informação de onde o trabalhador foi locado em cada trabalho, "position" com a informação de qual
era o cargo exercido pelo colaborador na empresa, e o "perfil_id" que é o identificador de cada funcionário.
    No arquivo  foi detectado várias linhas vazias apenas com o preenchimento do "perfil_id" como não possui nenhuma outra infor-
mação as linhas serão condiredadas nulas e serão deletadas do arquivo. AS linhas que possuíam alguma informação incompleta foi
inserido como "NA(não informado)" para que não haja nenhuma linha vazia e possa prejudicar a análise final dos dados.

    O arquivo "linguas" foi convertido para excel, assim foi separado automaticamente as colunas e foram identificadas as colunas
"idioma" com o idioma que o colaborador possui como habilidade, "proficiency" com a informação do nível que o funcionário possui 
no idioma em questão e a coluna "perfil_id" que identifica cada colaborador. Foi verificado muitas linhas em branco que foram
deletadas, foi verificado uma falha da conversão do arquivo e não tinha sido criada a coluna idioma, esta foi criada e todas as
ocorrências em linhas da palavra idioma que nao possuíam mais nenhuma informação foram deletadas também. Assim o arquivo foi pre-
parado para as análises.
```
